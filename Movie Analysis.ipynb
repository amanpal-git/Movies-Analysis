# Movies which have rating more than 8.5

```python
above_8 = df[df['rating']> 8.5 ]
above_8
```

### Checking for highest rated movies

```python
above_8['rating'].plot(kind='box')
```

# Plotting the graphs of the movies which have rating more than 8.5

```python
style.use('ggplot')
plt.figure(figsize=(20,25))
above_8=above_8.sort_values('rating')
sns.barplot(data=above_8, y=above_8['Movie-Title'], x=above_8['rating'], color='orangered')

plt.title("Movies with more than 8.5 IMDB rating",fontsize=50,fontweight='bold')

plt.xlabel("Rating", fontsize=45, fontweight='bold')
plt.ylabel("Movie Title", fontsize=45, fontweight='bold')  
plt.xticks(fontsize=40)
plt.yticks(fontsize=25)

for index, value in enumerate(above_8['rating']):
    plt.text(value, index,str(value),fontsize=25)

plt.show()
```

# Top 10 highest grossed movies

```python
rate_prof=df[['Movie-Title','rating','gross(M)']]
rate_prof
```

```python
rate_prof.sort_values('gross(M)',ascending=False,inplace=True)
rate_prof=rate_prof.head(10)
rate_prof
```

# Plotting the graphs of the top 10 highest grossed movies

```python
plt.figure(figsize=(7,7))
plt.title('Movies Collections',fontweight='bold')
plt.barh(rate_prof['Movie-Title'], rate_prof['gross(M)'],color='purple')
plt.xlabel('gross(M)',fontsize=15)
plt.ylabel('Movie-Title',fontsize=15)
for index, value in enumerate(rate_prof['gross(M)']):
    plt.text(value,index,str(value),fontsize=10)

plt.show()
```
